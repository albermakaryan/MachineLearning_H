Features:

- n_visits - number of visits of the website during the first 3 days.
- n_actions - number of actions/button clicks in the website during the first 3 days.
- unique_actions - number of unique actions in the website during the first 3 days.
- products_tried - number of products viewed during the first 3 days.
- products_added - number of products added to cart during the first 3 days.
- unique_products_category_tried - number of unique products’ categories viewed during the first 3 days.
- Unique_products_category_added - number of unique products’ categories added to cart during the first 3 days.
- saved_to_cart - if the user actually saved the purchases for later purchase during the first 3 days.
- comments_writen - how many unique comments the user did during the first 3 days.
- counts_of_buys - how many times the user actually paid during the first 3 days.

The boxplots above tells that almost every features has outliear values, which have negative influence on model accuracy. 
In the next step I am going to use sigmoid function to create a new dataframe with new value between [0,1] interval and the 
check if outlier problem is gone.

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

df_sigmoid = df.apply(lambda x: sigmoid(x) if len(np.unique(x)) > 2 else x)

df_sigmoid.head()
```

## Models 

```python
X_train,X_test, y_train,y_test = train_test_split(X,y, test_size=0.1, random_state=123)
```

## KNN

```python
knn_pip = Pipeline([
    ('scaler',StandardScaler()),
    ('knn', KNeighborsClassifier())
    ])

knn_pip.get_params()
```

```python
knn_params = {'knn__n_neighbors': list(range(1,30))}

knn_grid = GridSearchCV(knn_pip,
                       param_grid=knn_params,
                       cv=5)

```

```python
knn_grid.fit(X_train,y_train)
```

## Decision Tree

```python
tree_pip = Pipeline([
    ('scaler', StandardScaler()),
    ('tree', DecisionTreeClassifier())
])
tree_pip.get_params()
```

```python
tree_params = {"tree__max_depth":list(range(1,10)),
               "tree__min_samples_split":list(range(1,10))}

tree_grid = GridSearchCV(tree_pip,
                         tree_params,
                         cv=5)
```

```python
tree_grid.fit(X_train,y_train)
```

## Logistic Regression

```python
log_pip = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression())
])
log_pip.get_params()
```

```python
log_params = {"model__max_iter":[50,100,150],
              "model__tol":[0.0001,0.001,0.01]}

log_grid = GridSearchCV(log_pip,
                        log_params,
                        cv=5)
```

```python
log_grid.fit(X_train,y_train)
```

## Naive Bayes

```python
nb_pip = Pipeline([
    ('scaler', StandardScaler()),
    ('model', GaussianNB())
])
nb_pip.get_params()
```

```python
nb_params = {"model__priors":[0.001,0.01,0.1,0.5,0.7,0.9]}
nb_grid = GridSearchCV(nb_pip,
                       nb_params,
                       cv=5)
```

```python
nb_grid.fit(X_train,y_train)
```
